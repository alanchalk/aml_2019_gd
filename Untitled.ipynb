```python
import numpy as np
import gd
```

```python
def fn_loss(x):
    return (x + 2)**2
    
def fn_grad(x):
    return 2*(x + 2)
```

```python
solver = gd.gd_pv_1d(fn_loss = fn_loss, fn_grad = fn_grad)
```

```python
solver.find_min(x_init = 4, n_iter = 10000, eta = 0.001, tol = 1e-5)
```

```python
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline

font = {'size': 18}
matplotlib.rc('font', **font)

cmap = matplotlib.cm.get_cmap('viridis')
lst_loss_log = np.log(solver.loss_path)
normalize = matplotlib.colors.Normalize(vmin=min(lst_loss_log), 
                                        vmax=max(lst_loss_log))

colours = [cmap(normalize(value)) for value in lst_loss_log]

fig = plt.figure(figsize=(10, 8))
ax1 = fig.add_subplot(1, 1, 1)
ax1.scatter(solver.x_path, solver.loss_path, color=colours)
ax1.set_xlabel('x')
ax1.set_ylabel('loss')
```
